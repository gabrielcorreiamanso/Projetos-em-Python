## Pergunta de Negócio 2:

### Qual o Total de Vendas Por Data do Pedido?

Demonstre o resultado através de um gráfico de barras.

## Pergunta de Negócio 5:

### Qual Segmento Teve o Maior Total de Vendas?

Demonstre o resultado através de um gráfico de pizza.

```python
df_dsa_p5 = df_dsa.groupby('Segmento')['Valor_Venda'].sum().reset_index().sort_index()
```

```python
df_dsa_p5
```

```python
plt.figure(figsize = (16,6))
plt.pie(df_dsa_p5['Valor_Venda'], labels=df_dsa_p5['Segmento'])
plt.title('As 10 Cidades com Maior Total de Vendas')
plt.show()
```

```python
# Função para converter os dados em valor absoluto
def autopct_format(values): 
    def my_format(pct): 
        total = sum(values) 
        val = int(round(pct * total / 100.0))
        return ' $ {v:d}'.format(v = val)
    return my_format
```

```python
# Plot

# Tamanho da figura
plt.figure(figsize = (16, 6))

# Gráfico de pizza
plt.pie(df_dsa_p5['Valor_Venda'], 
        labels = df_dsa_p5['Segmento'],
        autopct = autopct_format(df_dsa_p5['Valor_Venda']),
        startangle = 90)

# Limpa o círculo central
centre_circle = plt.Circle((0, 0), 0.82, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Labels e anotações
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(df_dsa_p5['Valor_Venda']))), xy = (-0.25, 0))
plt.title('Total de Vendas Por Segmento')
plt.show()
```

## Pergunta de Negócio 7 (Desafio Nível Júnior):

Os gestores da empresa estão considerando conceder diferentes faixas de descontos e gostariam de fazer uma simulação com base na regra abaixo:

- Se o Valor_Venda for maior que 1000 recebe 15% de desconto.
- Se o Valor_Venda for menor que 1000 recebe 10% de desconto.

### Quantas Vendas Receberiam 15% de Desconto?

```python
df_dsa.head(10)
```

```python
# Criar uma nova coluna 'Tipo_Desconto' para indicar o tipo de desconto
df_dsa['Tipo_Desconto'] = ''
```

```python
# Iterar sobre as linhas do DataFrame e determinar o tipo de desconto
for index, row in df_dsa.iterrows():
    # Verificar a condição e determinar o tipo de desconto
    if row['Valor_Venda'] >= 1000:
        tipo_desconto = '15% de desconto'  # Define o tipo de desconto como '15% de desconto'
    else:
        tipo_desconto = '10% de desconto'  # Define o tipo de desconto como '20% de desconto'
    
    # Atribuir o tipo de desconto à nova coluna
    df_dsa.at[index, 'Tipo_Desconto'] = tipo_desconto
```

```python
df_dsa['Valor_Com_Desconto'] = df_dsa['Valor_Venda']
```

```python
for index, row in df_dsa.iterrows():
    # Verificar a condição e aplicar o desconto apropriado
    if row['Valor_Venda'] >= 1000:
        df_dsa.at[index, 'Valor_Com_Desconto'] *= 0.85  # Subtrai 10% do valor original
    else:
        df_dsa.at[index, 'Valor_Com_Desconto'] *= 0.9  # Subtrai 20% do valor original
```

```python
df_dsa.query("Tipo_Desconto == '15% de desconto'").head(10)
```

```python
df_dsa.groupby('Tipo_Desconto')['Valor_Venda'].count().reset_index()
```

## Pergunta de Negócio 9 (Desafio Nível Master Ninja):

### Qual o Média de Vendas Por Segmento, Por Ano e Por Mês?

Demonstre o resultado através de gráfico de linha.

```python
df_dsa_p9 = df_dsa.groupby(['Segmento', 'Ano', 'Mês'])['Valor_Venda'].agg([np.sum, np.mean, np.median])
```

```python
df_dsa_p9.head()
```

```python
sns.set()
```

```python
# Plotando o gráfico de linha
plt.figure(figsize=(18, 6))
for segmento, dados_segmento in df_dsa_p9.groupby('Segmento'):
    plt.plot(dados_segmento['Ano'].astype(str) + '-' + dados_segmento['Mês'].astype(str), dados_segmento['sum'], label=f'{segmento} - Sum')

# Adicionando legendas e rótulos aos eixos
plt.xlabel('Ano e Mês')
plt.ylabel('Valor de Venda (Soma)')
plt.title('Evolução das Vendas ao Longo do Tempo')
plt.legend()
plt.grid(False)  # Removendo as linhas de grade

# Rotacionando os rótulos do eixo x para melhor visualização
plt.xticks(rotation=45)

# Mostrando o gráfico
plt.tight_layout()
plt.show()
```
